```python
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Load dataset
df = pd.read_csv("train.csv", encoding="utf-8")

# Ensure correct column names
df = df[['comment_text', 'toxic']]

# Handle missing values
df.dropna(subset=['comment_text', 'toxic'], inplace=True)

# Convert 'toxic' labels to integers (1 = toxic, 0 = not toxic)
df['toxic'] = df['toxic'].astype(int)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['comment_text'], df['toxic'], test_size=0.2, random_state=42)

# Vectorize the text data using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', max_features=10000, ngram_range=(1,2))
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train a Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_tfidf, y_train)

# Make predictions on test data
y_pred = model.predict(X_test_tfidf)

# Print evaluation metrics
print("\nModel Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Save the trained model and vectorizer using pickle
with open('toxic_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

with open('vectorizer.pkl', 'wb') as vectorizer_file:
    pickle.dump(vectorizer, vectorizer_file)

print("\nModel training complete and saved!")

# Function to make a prediction
def predict_toxicity(text):
    """ Predict whether a given text is Toxic or Not Toxic """
    with open('vectorizer.pkl', 'rb') as vectorizer_file:
        vectorizer = pickle.load(vectorizer_file)  # Load vectorizer
    with open('toxic_model.pkl', 'rb') as model_file:
        model = pickle.load(model_file)  # Load model

    vectorized_text = vectorizer.transform([text])

    # Check if the input text has any nonzero values in the vectorized form
    if np.sum(vectorized_text) == 0:
        return "Not Toxic"  # If the text is not in the dataset, return "Not Toxic"

    prediction = model.predict(vectorized_text)
    return "Toxic" if prediction[0] == 1 else "Not Toxic"
```
